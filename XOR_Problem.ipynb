```python
# Implement a Multilayer Perceptron MLP and train it using the back propogation alogorithm
# Train an MLP on XOR problem and demonstrate gradient based learning and the use of hidden units
```

```python
# Import the necessary libraries
import numpy as np
```

```python
# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Define XOR input and expected output
X= np.array([[0,0],[0,1],[1,0],[1,1]])
y= np.array([[0],[1],[1],[0]])

# Set random seed for consistent results
np.random.seed(0)
```

```python
# Initialize weights randomly husing uniform distribution
W1= np.random.uniform(-1,1,(2,4))
W2= np.random.uniform(-1,1,(4,1))
lr= 0.1
```

```python
# Training the network (backpropogation)
for epoch in range(20000):
  # Forward Prop
  hidden_layer= sigmoid(X@W1)
  output_layer= sigmoid(hidden_layer@W2)

  # Compute error
  error= y-output_layer

  # Backprop
  d_output= error*sigmoid_derivative(output_layer)
  d_hidden= (d_output@W2.T)*sigmoid_derivative(hidden_layer)

  # Update weight
  W2+= hidden_layer.T@d_output*lr
  W1+= X.T@d_hidden*lr

  if epoch%5000==0:
    print(f"Epoch {epoch}-> Loss: {np.mean(np.abs(error)):.4f}")
```

```python
# Print Final Predictions
print("Final Output after training: ")
print(output_layer.round(3))
```
